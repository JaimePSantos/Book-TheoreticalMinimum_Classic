```python
import networkx as nx
import numpy as np
from matplotlib import pyplot as plt
from math import sqrt, ceil, pow
```

# r1 = (cos wt, e^wt)

```python
def r1_x(time,w):
    rList = []
    vList = []
    aList = []
    allList = []
         
    for t in time:
        rList.append(np.cos(w*t))
        vList.append(-w*np.sin(w*t))
        aList.append(-w**2 * np.cos(w*t))
    allList.append(rList)
    allList.append(vList)
    allList.append(aList)
    return allList

time = np.linspace(0,1.5,100)
p1_x = r1_x(time,2*np.pi)

for p in p1_x:
    plt.plot(time,p,label="bla")
plt.legend()
plt.show()
```

```python
def r1_y(time,w):
    rList = []
    vList = []
    aList = []
    allList = []
         
    for t in time:
        rList.append(np.exp(w*t))
        vList.append(w*np.exp(w*t))
        aList.append(w**2 * np.exp(w*t))
    allList.append(rList)
    allList.append(vList)
    allList.append(aList)
    return allList

time = np.linspace(0,1.5,100)
p1_y = r1_y(time,2*np.pi)

for p in p1_y:
    plt.plot(time,p,label="bla")
plt.legend()
plt.show()
```

# r2 = (cos(wt-phi),sin(wt-phi))

```python
def r2_x(time,w,phi):
    rList = []
    vList = []
    aList = []
    allList = []
         
    for t in time:
        rList.append(np.cos(w*t - phi))
        vList.append(-w*np.sin(w*t-phi))
        aList.append(-w**2 * np.cos(w*t-phi))
    allList.append(rList)
    allList.append(vList)
    allList.append(aList)
    return allList

time = np.linspace(0,1,100)
phi = np.pi/2
w = 2*np.pi
p2_x = r2_x(time,w,phi)

for p in p2_x:
    plt.plot(time,p,label="bla")
plt.legend()
plt.show()
```

```python
def r2_y(time,w,phi):
    rList = []
    vList = []
    aList = []
    allList = []
         
    for t in time:
        rList.append(np.sin(w*t - phi))
        vList.append(w*np.cos(w*t-phi))
        aList.append(-w**2 * np.sin(w*t-phi))
    allList.append(rList)
    allList.append(vList)
    allList.append(aList)
    return allList

time = np.linspace(0,1,100)
phi = np.pi/2
w = 2*np.pi
p2_y = r2_y(time,w,phi)

for p in p2_y:
    plt.plot(time,p,label="bla")
plt.legend()
plt.show()
```

# r3 = (c*cos^3(t),c*sin^3(t))

```python
def r3_x(time,c):
    rList = []
    vList = []
    aList = []
    allList = []
         
    for t in time:
        rList.append(c*np.cos(t)**3)
        vList.append(-3*c*np.cos(t)**2 * np.sin(t))
        aList.append(-3*c*np.cos(t)*(np.cos(t)**2 - 2*np.sin(t)**2))
    allList.append(rList)
    allList.append(vList)
    allList.append(aList)
    return allList

time = np.linspace(0,2,100)
c = 1
p3_x = r3_x(time,c)

for p in p3_x:
    plt.plot(time,p,label="bla")
plt.legend()
plt.show()
```

```python
def r3_y(time,c):
    rList = []
    vList = []
    aList = []
    allList = []
         
    for t in time:
        rList.append(c*np.sin(t)**3)
        vList.append(3*c*np.sin(t)**2 * np.cos(t))
        aList.append(-3*c*np.sin(t)*(np.sin(t)**2 - 2*np.cos(t)**2))
    allList.append(rList)
    allList.append(vList)
    allList.append(aList)
    return allList

time = np.linspace(0,2,100)
c = 1
p3_y = r3_y(time,c)

for p in p3_y:
    plt.plot(time,p,label="bla")
plt.legend()
plt.show()
```

# r3 = (c*(t-sin(t),c*(1-cos(t)))

```python
def r4_x(time,c):
    rList = []
    vList = []
    aList = []
    allList = []
         
    for t in time:
        rList.append(c*(t-np.sin(t)))
        vList.append(c*(1-np.cos(t)))
        aList.append(c*np.sin(t))
    allList.append(rList)
    allList.append(vList)
    allList.append(aList)
    return allList

time = np.linspace(0,1,100)
c = 0.5
p4_x = r4_x(time,c)

for p in p4_x:
    plt.plot(time,p,label="bla")
plt.legend()
plt.show()
```
